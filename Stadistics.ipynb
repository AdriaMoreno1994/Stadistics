### Exercici 1 

Calcul la moda, la mediana la desviació estandard i la mitjana artimetica d'un atribut



```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
df = pd.read_csv(r'C:/Users/adria/tasca 7/summary.csv')
```

```python
df.head()
```

```python
moda_gamesPlayed = df['GamesPlayed'].mode()[0]
print('la moda de partits jugats es :',moda_gamesPlayed)
```

```python
mediana_gamesPlayed = df['GamesPlayed'].median()
print('La mediana de partits jugats és:', mediana_gamesPlayed)
```

```python
desviacio_standard = df['GamesPlayed'].std()
```

```python
print('la desviació estandard de partits jugats és:',round(desviacio_standard,3))
```

```python
mitjana = df['GamesPlayed'].mean()
print('la mitjana artimetica de partits jugats és :',round(mitjana,3))
```

selecciona dos atributs i calcula la correlació
